```python
# lets create a list with 10 yearly salaries for a Senior Data Analyst job
```

```python
import random

salary = [random.randint(100_000 , 150_000) for num in range (10)]
salary
```

```python
#Now we'll convert this list into a NumPy array.
salary_array = np.array(salary)
salary_array
```

```python
np.sum(salary_array)
```

```python
#Calculate the product of the elements
np.prod(salary_array)
```

```python
# Calculates the cumulative sum of elements
np.cumsum(salary_array)
```

```python
# Calculates the cumulative product of elements
np.cumprod(salary_array)
```

```python
# Calculate the average salary
np.mean(salary_array)
```

```python
# find the median salary
np.median(salary_array)
```

```python
# calculate the variance
np.var(salary_array , ddof=1)   #ddof=1 for sample variance
```

```python
# Calculate the standard deviation
np.std(salary_array , ddof=1 )            ## ddof=1 for sample standard deviation
```

```python
import time

size = 1000000
```

```python
# python list
l1 = range(size)
l2 = range(size)

start = time.time()
result = [(x+y) for x,y in zip(l1,l2)]
print("python list took: ", (time.time()-start)*1000)
```

```python
# numpy array
a1 = np.arange(size)
a2 = np.arange(size)

start = time.time()
result = a1 + a2
print("numpy took: ", (time.time()-start)*1000)
```

```python
# 3rd way
for x in np.nditer(a, order= 'C'):    # "c" order go by rows to row, "fortran" order goes by column to column
  print(x)
```

```python
for x in np.nditer(a, order= 'F', flags=['external_loop']):
  print(x)
```
