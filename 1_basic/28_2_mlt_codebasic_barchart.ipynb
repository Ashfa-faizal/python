```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
company = ['GOOGLE','AMZN','MSFT','FB']
revenue = [90,136,89,27]
profit = [40,3,23,23]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos, company)
plt.ylabel("Revenue(bill)")
plt.title("US tech stocks")
plt.bar(xpos-0.2, revenue, color='skyblue' ,width=0.4, label='Revanue')  # by xpos-0.2 we are moving x exis(company)

plt.bar(xpos+0.2, profit , color ='grey',width=0.4, label='profit')

plt.legend()
plt.show()
```

```python
plt.yticks(xpos, company)
bar_height = 0.4

plt.title("US tech stocks")
plt.barh(xpos-0.2, revenue, color='skyblue' , height=bar_height, label='Revanue')

plt.barh(xpos+0.2, profit , color ='grey', height=bar_height, label='profit')

plt.legend()
plt.show()
```
