```python
import matplotlib.pyplot as plt
```

```python
blood_sugar = [118,78,119,79,65,84,86,94,104,99,104,153,80,162,52]
# histrogram take onle single dimention array , y axis contains frequency
# by default it'll plot 10 bin
# r width is relative width of a bar compared to bin size max_value is 1
```

```python
plt.hist(blood_sugar, bins=[80,100,125,160], rwidth= 0.95)
```

```python
women_blood_sugar = [85, 92, 88, 101, 95, 89, 110, 93, 97, 84, 60, 99, 86, 160]
men_blood_sugar   = [91, 96, 104, 98, 87, 100, 94, 107, 85, 92, 108, 90, 60, 159]

plt.xlabel('Blood Sugar Level (mg/dL)')
plt.ylabel('Number of Patients')
plt.title('Blood Sugar Analysis')
plt.hist([women_blood_sugar,men_blood_sugar], bins=[80,100,125,160], rwidth= 0.95, color=['green','orange'],
         label=['women','men'])
plt.legend()
```
