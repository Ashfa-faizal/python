#plotting weather data

```python
import matplotlib.pyplot as plt
%matplotlib inline     #it can render the chart fine in jupyter notebook
```

```python
x = [1,2,3,4,5,6,7]     # number of weekdays
y = [34,56,54,32,65,49,23]       #temperature information
```

```python
plt.xlabel("day")
plt.ylabel("Temperature")
plt.title("Daily Temperature")
plt.plot(x, y, color='green', marker='*', linestyle='dashed',
     linewidth=2, markersize=12)             #for color and marker u can simply type "g*"
```

# plotting multiple data

```python
import matplotlib.pyplot as plt
```

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,47,48,45,47,51]
min_t = [43,42,38,37,36,34,34]
avg_t = [48,49,43,44,40,43,49]
```

```python
plt.xlabel('days')
plt.ylabel('temperature')
plt.title('weather')
plt.plot(days,max_t, label = 'max')
plt.plot(days,min_t, label = 'min')
plt.plot(days,avg_t, label = 'avg')

plt.legend(loc="best", shadow=True , fontsize="large")   # mtl best will deside where to set the legend

plt.grid()
```
