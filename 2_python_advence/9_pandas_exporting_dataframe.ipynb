```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\HP\Desktop\ASHFA\data_jobs.csv")

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# in data analyst job
df_DA = df[(df['job_title_short'] == 'Data Analyst')].copy()
df_DA['job_posted_month'] = df_DA['job_posted_date'].dt.strftime('%b')

months = df_DA['job_posted_month'].unique()
df_DA_month = {}
for month in months:
    df_DA_month[month] = df_DA[df_DA['job_posted_month'] == month].copy()
```

```python
df_q1 = pd.concat([df_DA_month['Jan'],df_DA_month['Feb'],df_DA_month['Mar']],ignore_index=True)
df_q1
```
