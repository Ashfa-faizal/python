* histograms are great at showing the distribution of values

```python
import pandas as pd
import matplotlib.pyplot as plt
import ast

df = pd.read_csv(r"C:\Users\HP\Desktop\ASHFA\data_jobs.csv")

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()
df_DA_US = df_DA_US.dropna(subset=['salary_year_avg'])
#When you want to drop rows that have missing values only in certain columns,
#instead of the entire DataFrame, you use the subset argument.

df_DA_US.head(2)
```

```python
df_DA_US['salary_year_avg'].plot(kind='hist' , bins=30, edgecolor ='black')
```

short tail into 300000 

```python
df_DA_US['salary_year_avg'].plot(kind='hist' , bins=40, edgecolor ='black')
plt.xlim(0 , 250000)

# formatting number value to doller sign

plt.title('Distribution of United States Data Analyst Yearly Salaries')
plt.xlabel('Yearly Salary')
plt.ylabel('Number of Jobs')
plt.show()
```
