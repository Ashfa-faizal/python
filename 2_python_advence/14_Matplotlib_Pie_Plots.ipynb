```python
import pandas as pd
import matplotlib.pyplot as plt
import ast

df = pd.read_csv(r"C:\Users\HP\Desktop\ASHFA\data_jobs.csv")

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie')
plt.title('work from home status')
plt.ylabel("")
plt.show()
```

```python
df['job_title_short'].value_counts().plot(kind='pie')
plt.title('titles of jobs')
plt.ylabel("")
plt.show()
```

```python
# we can't compare to each other like which one higher/ lower ,but in bar chart we easily compare

df['job_title_short'].value_counts().plot(kind='barh')
plt.title('titles of jobs')
plt.gca().invert_yaxis()
plt.ylabel("")
plt.show()
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie' , startangle=90 , autopct = '%1f%%')
# startangle 0 is always in the right side   # in autopct we want 1 decimal place '%1.1f'  ( f is floating point number)
plt.title('work from home status')
plt.ylabel("")
plt.show()
```

```python
fig, ax = plt.subplots(1,3)

dict_column = {
    'job_work_from_home': 'work from home',
    'job_no_degree_mention': 'job degree rag',
    'job_health_insurance': 'health insurance mention'
}

for i,(column,title) in enumerate(dict_column.items()):
    ax[i].pie(df[column].value_counts(), startangle=90 , autopct = '%1.1f%%', labels=['False','True'])   # be aware when u giving a labe;)
    ax[i].set_title(title)

plt.show
```
