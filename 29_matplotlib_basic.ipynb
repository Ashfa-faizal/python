```python
# counting how many job posting are happening relative to that job posted date
date_counts = df.job_posted_date.value_counts()
date_counts
```

```python
type(date_counts)         # now it's series
```

```python
# date and time(x) are index, these value on the right hand side(y) is core potion of the series
plt.plot(date_counts.index, date_counts)
```

```python
# it's mess because the index is not in the neumerical order
date_counts = date_counts.sort_index()
date_counts.head()
```

```python
plt.plot(date_counts.index, date_counts)
```

```python
# it's still mess bcs of  the time second different
# so we need to aggregate by day / week / month / quarter
```

```python
# let's aggerate by month  , first add month column
df['job_posted_month'] = df.job_posted_date.dt.month
```

```python
monthly_counts = df.job_posted_month.value_counts()
monthly_counts = monthly_counts.sort_index()
monthly_counts
```

```python
plt.plot(monthly_counts.index, monthly_counts)      # we can access value inside by providing  (monthly_counts.value)
```

#series vs dataframe


1. Pandas Series
* A Series is a one-dimensional labeled array.
* Think of it like a single column of data — like a list with labels (index).
2. Pandas DataFrame
* A DataFrame is a two-dimensional labeled data structure.
* Think of it like a table — with rows and columns.

```python
series = pd.Series([10, 20, 30, 40], index= [ 'a','b','c','d'])
```
